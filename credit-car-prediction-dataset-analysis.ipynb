```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv('/kaggle/input/car-price-prediction-dataset/test_data.csv')
train = pd.read_csv('/kaggle/input/car-price-prediction-dataset/train_data.csv')
```

```python
print("Test Dataset Header:")
print(test.head())

print("Train Dataset Header:")
print(train.head())
```

```python
print("Test Dataset Info:")
print(test.info())

print("Train Dataset Info:")
print(train.info())
```

```python
print("Sum of Null Values in Test Dataset:")
print(test.isnull().sum())

print("Sum of Null Values in Train Dataset:")
print(train.isnull().sum())
```

```python
test['Job title'] = test['Job title'].fillna('Unknown')
train['Job title'] = train['Job title'].fillna('Unknown')
```

```python
print("Sum of Null Values in Test Dataset:")
print(test.isnull().sum())

print("Sum of Null Values in Train Dataset:")
print(train.isnull().sum())
```

```python
# Convert the columns to integer type
test['Family member count'] = test['Family member count'].astype(int)
test['Account age'] = test['Account age'].astype(int)

train['Family member count'] = train['Family member count'].astype(int)
train['Account age'] = train['Account age'].astype(int)
```

```python
test = test.drop_duplicates()
train = train.drop_duplicates()
```

```python
# Distribution of Income
plt.figure(figsize=(10, 6))
sns.histplot(train['Income'], bins=30, kde=True)
plt.title('Distribution of Income in Train Dataset')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of Gender distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=train)
plt.title('Gender Distribution in Train Dataset')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot of 'Has a car' distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Has a car', data=train)
plt.title('Car Ownership in Train Dataset')
plt.xlabel('Has a car')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot of 'Has a property' distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Has a property', data=train)
plt.title('Property Ownership in Train Dataset')
plt.xlabel('Has a property')
plt.ylabel('Count')
plt.show()
```

```python
# Pair plot to see the relationship between numerical variables
sns.pairplot(train[['Income', 'Age', 'Employment length', 'Family member count', 'Account age']])
plt.show()
```

```python
# Heatmap of correlation matrix
plt.figure(figsize=(12, 8))

# Selecting only the numerical columns
numerical_columns = train.select_dtypes(include=[np.number])
correlation_matrix = numerical_columns.corr()

# Creating the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Train Dataset')
plt.show()
```

```python
# Box plot of Income vs. Employment Status
plt.figure(figsize=(12, 6))
sns.boxplot(x='Employment status', y='Income', data=train)
plt.title('Income vs. Employment Status in Train Dataset')
plt.xlabel('Employment Status')
plt.ylabel('Income')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot of Income vs. Education Level
plt.figure(figsize=(12, 6))
sns.boxplot(x='Education level', y='Income', data=train)
plt.title('Income vs. Education Level in Train Dataset')
plt.xlabel('Education Level')
plt.ylabel('Income')
plt.xticks(rotation=45)
plt.show()
```
